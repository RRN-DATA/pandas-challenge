```python
%matplotlib notebook
# Dependencies and Setup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.DataFrame(pd.read_csv(file_to_load))
```

```python
#Calculating the total number of players in purchase_data
tot_players = purchase_data['SN'].nunique()
players_summary = pd.DataFrame({"Total Players": [tot_players]})
players_summary
```

```python
#Obtaining number of unique items
num_uniq_items = purchase_data['Item ID'].nunique()
#Calculating Average Price
avg_price = round(purchase_data['Price'].mean(), 2)
#Obtaining number of purchases
numb_purchases = purchase_data['Purchase ID'].nunique()
#Calculating total revenue
tot_revenue = purchase_data['Price'].sum()
#Creating a dataframe
df_summary = pd.DataFrame({"Number Of Unique Items": [num_uniq_items],
                              "Average Price": "$"+str(avg_price),
                              "Number of Purchases": [numb_purchases],
                              "Total Revenue": "$"+str(tot_revenue)})
df_summary
```

```python
#Percentage count of male player
series = pd.Series(purchase_data['Gender'])
male_count = series.str.count("Male").sum()
female_count = series.str.count("Female").sum()
other_count = series.str.count("Other / Non-Disclosed").sum()
gender_total = purchase_data['Gender'].count()
male_count_per = (male_count/gender_total)*100
male_count_per = male_count_per.round(2)
female_count_per = (female_count/gender_total)*100
female_count_per = female_count_per.round(2)
other_count_per = (other_count/gender_total)*100
other_count_per = other_count_per.round(2)
gender_summary = pd.DataFrame({"Total Count": [male_count, female_count, other_count],
                              "Percentage Of Players": [str(male_count_per)+"%", str(female_count_per)+"%", str(other_count_per)+"%"]})
gender_summary.index = ['Male', 'Female', 'Other / Non-Disclosed']
gender_summary
```

```python
purchase_drop_dupli = purchase_data.drop_duplicates(subset="Purchase ID",keep="first")

pur_counts=purchase_data["Gender"].value_counts()
pur_counts_perperson=purchase_drop_dupli["Gender"].value_counts()
tot_purchases=purchase_data.groupby("Gender")["Price"].sum()
avg_pur_price=tot_purchases/pur_counts
tot_purchase_per_person=round((tot_purchases/pur_counts_perperson),2)

pd.options.display.float_format = '${:,.2f}'.format
purchase_analysis =pd.DataFrame({"Purchase Count":pur_counts,"Average Purchase Price": avg_pur_price,"Total Purchase Value": tot_purchases,"Avg Total Purchase per Person": tot_purchase_per_person}).sort_index(axis=0)
purchase_analysis
```

```python
# Establish Bins
purchase_data = pd.read_csv(file_to_load)
age_bins = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
age_bin_build = [0,10,15,20,25,30,35,40, 50]
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], age_bin_build ,right=False,labels=age_bins,include_lowest=True)
purchase_unique = purchase_data.drop_duplicates(subset="SN",keep="first")
```

```python
#Calculate Range of Age, Total and Percentage
age_range=purchase_unique["Age Range"].value_counts()
percent_players=age_range/len(purchase_unique)

#preparing a dataframe
pur_age_demog =pd.DataFrame({"Total Count":age_range,"Percentage of Players": percent_players}).sort_index(axis=0)
pur_age_demog
```

```python
#Adding Style format
pur_age_demog_format={'TotalCount':'{:}','Percentage of Players':'{:.2%}'}
formatted_purchase_age_demographics=pur_age_demog.style.format(pur_age_demog_format)
formatted_purchase_age_demographics
```

```python
purchase_unique = purchase_data.drop_duplicates(subset="Purchase ID",keep="first")

#Find Total PurchaseCounts
purc_counts=purchase_data["Age Range"].value_counts()
#Apply Counts as 1 for Duplicate Person, by using unique person data set
purc_counts_perperson=purchase_unique["Age Range"].value_counts()
age_group_df=purchase_data.groupby("Age Range")
#Total Purchase Value , Average Purchase Price, #Total Purchase Value , Average Purchase Price, 
tot_purcs=age_group_df["Price"].sum()
avg_purcs=round((tot_purcs/purc_counts),2)
tot_purcs_pp=round((tot_purcs/purc_counts_perperson),2)
#prepare a dataframe
pd.options.display.float_format = '${:,.2f}'.format
Purchase_Analysis =pd.DataFrame({"Purchase Count":purc_counts,"Average Purchase Price": avg_purcs,"Total Purchase Value": tot_purcs,"Avg Total Purchase per Person": tot_purcs_pp}).sort_index(axis=0)
Purchase_Analysis
```

```python
purchaser_counts = purchase_data.groupby("SN")["SN"].count()
tot_pur_perperson =purchase_data.groupby("SN")["Price"].sum()
avg_pur_perperson = tot_pur_perperson/purchaser_counts

top_spenders =pd.DataFrame({"Purchase Count":purchaser_counts,"Average Purchase Price": avg_pur_perperson,"Total Purchase Value": tot_pur_perperson,})
top_spenders=top_spenders.sort_values(['Purchase Count'], ascending=[0])
top_spenders.head()
```
